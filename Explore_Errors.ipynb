```python
sgsm = pd.read_json("results/gpt-35-turbo/samples_sgsm_2024-06-10T23-01-01.090319.json", orient='records', lines=True)
sgsm = sgsm.iloc[2047:]
sgsm_errors = sgsm[sgsm['exact_match']==0]
sgsm_errors.head()
```

```python
def print_errors(df, num_qs = 10):
    import random
    for i in range(num_qs):
        idx = random.randint(0,len(df)-1)
        question = df.iloc[idx]['doc']['question']
        answer = df.iloc[idx]['doc']['answer']
        model_answer = df.iloc[idx]['filtered_resps'][0]
        print(f"Question: {question}\nGround Truth: {answer}\nModel Answer: {model_answer}\n\n")
```

```python
print_errors(sgsm_errors)
```

```python
print_errors(gsm8k_errors)
```
